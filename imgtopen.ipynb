```python
import cv2
from google.colab.patches import cv2_imshow
```

```python
image = cv2.imread("dogrose.jpg")
cv2_imshow(image)
cv2.waitKey(0)
```

```python
gr_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(gr_image)
cv2.waitKey(0)
```

```python
invert = 255 - gr_image
cv2_imshow(invert)
cv2.waitKey()
```

```python
blurr = cv2.GaussianBlur(invert, (21, 21), 0)
```

```python
invert = 255 - blurr
pencil_sketch = cv2.divide(gr_image, invert, scale=256.0)
cv2_imshow(pencil_sketch)
cv2.waitKey(0)
```

```python
cv2_imshow(image)
cv2_imshow(pencil_sketch)
cv2.waitKey(0)
```
